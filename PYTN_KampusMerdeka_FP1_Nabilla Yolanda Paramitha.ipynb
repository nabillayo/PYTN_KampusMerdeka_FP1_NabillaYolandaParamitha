## Data Preprocessing

### Encoding Label

```python
# Menampilkan informasi mengenai data
data_fin.info()
```

```python
# Encoding kolom dengan tipe data object menjadi data kategorik 
le = preprocessing.LabelEncoder()

data_fin['agencies'] = le.fit_transform(data_fin.agencies)
data_fin['service_type'] = le.fit_transform(data_fin.service_type)
data_fin['start'] = le.fit_transform(data_fin.start)
data_fin['finish'] = le.fit_transform(data_fin.finish)
data_fin['weather'] = le.fit_transform(data_fin.weather)

data_fin.head()
```

```python
# Menampilkan data unik dari kolom service_type
data_fin['service_type'].unique()
```

```python
# visualisasi untuk melihat korelasi menggunakan heatmap
plt.figure(figsize=(14,12))
sns.heatmap(data_fin.corr(), vmin=-1, vmax=1, annot=True)

plt.show()
```

<b> Kesimpulan : </b>

- Dari heat map di atas, dapat dilihat nilai korelasi antar variabel dengan tarif taxi. Dari ke-11 fitur, hanya fitur <b> distance </b> dan <b> surge_multiplier </b> yang memiliki korelasi positif dengan tarif. 
- Fitur <b> service_type </b> juga menunjukkan adanya korelasi, hanya saja korelasi yang terjadi merupakan korelasi negatif sehingga fitur service_type juga dimasukkan ke dalam model.
- Sementara itu, fitur lain yang memiliki korelasi yang kecil dengan harga, tidak dimasukkan ke dalam model.

```python
# Memilih kolom yang memiliki korelasi dengan tarif taxi
data_regresi = data_fin[['distance', 'surge_multiplier', 'service_type', 'price']]
data_regresi.head()
```

```python
# Mendefinisikan variabel x dan variabel y
X =  data_regresi.drop(columns=['price'])
y = data_regresi['price']
# Memisahkan data training dan data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Defining Model

```python
# Membuat variabel model sebagai instance dari LinearRegression
model = LinearRegression()
```

### Training Model

```python
model.fit(X_train, y_train)
```

```python
# Melihat nilai intercept
model.intercept_
```

```python
# Melihat nilai koefisien
model.coef_
```

```python
# Prediksi data testing
y_pred = model.predict(X_test)
y_pred
```

<b><i> Insight : </i></b>

Dari proses pemodelan di atas diperoleh model sebagai berikut:

Y = -0.36099417983663074 + 2.7909668 X1 + 18.98447144 X2 - 1.53077175 X3

dengan

Y = price

X1 = distance

X2 = surge_multiplier

X3 = service_type

### Model Evaluation

```python
# Nilai R-Squared
r2_score(y_test, y_pred)
```

<b><i> Insight : </i></b>

Terlihat nilai R-Squared sebesar 0.5 yang artinya variasi tarif taxi dapat dijelaskan oleh distance, surge_multiplier dan service_type sebesar 50%. Sedangkan, sebanyak 50% dijelaskan oleh faktor lain.

### Visualisasi Data Asli dengan Data Prediksi
